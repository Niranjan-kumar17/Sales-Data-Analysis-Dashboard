```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Random seed
random.seed(42)
np.random.seed(42)

# Number of Records
num_records = 5000

# Lists
states = ["Delhi", "Maharashtra", "Karnataka", "Bihar", "Gujarat", "Madhya Pradesh"]

cities = {
    "Delhi": ["New Delhi"],
    "Maharashtra": ["Mumbai", "Pune", "Nagpur"],
    "Karnataka": ["Bengaluru", "Mysuru"],
    "Bihar": ["Patna", "Darbhanga"],
    "Gujarat": ["Ahmedabad", "Surat"],
    "Madhya Pradesh": ["Indore", "Bhopal"]
}

regions = {
    "Delhi": "North",
    "Maharashtra": "West",
    "Karnataka": "South",
    "Bihar": "East",
    "Gujarat": "West",
    "Madhya Pradesh": "Central"
}

segments = ["Consumer", "Corporate", "Home Office"]

categories = {
    "Furniture": ["Chair", "Table", "Sofa", "Bookcase"],
    "Technology": ["Laptop", "Printer", "Keyboard", "Mouse"],
    "Office Supplies": ["Notebook", "Pen", "Paper", "Marker"]
}

payment_modes = ["Cash", "UPI", "Credit Card", "Debit Card", "Net Banking"]

ship_modes = ["Standard Class", "Second Class", "First Class", "Same Day"]

customer_names = [
    "Rahul","Amit","Neha","Priya","Ankit","Rohan","Sneha",
    "Pooja","Karan","Anjali","Vikas","Nisha","Saurabh",
    "Ritika","Arjun","Riya","Deepak","Shweta","Akash","Simran"
]

# Start Date
start_date = datetime(2023, 1, 1)

data = []

for i in range(1, num_records + 1):

    state = random.choice(states)
    city = random.choice(cities[state])

    category = random.choice(list(categories.keys()))
    sub_category = random.choice(categories[category])

    sales = round(np.random.uniform(200, 10000), 2)
    quantity = random.randint(1, 10)
    discount = random.choice([0, 5, 10, 15, 20])

    profit = round(sales * np.random.uniform(-0.05, 0.30), 2)

    order_date = start_date + timedelta(days=random.randint(0, 730))
    ship_date = order_date + timedelta(days=random.randint(1, 7))

   # Calculate Delivery Days
    delivery_days = (ship_date - order_date).days

# Calculate Profit Margin
    profit_margin = round((profit / sales) * 100, 2)

    data.append({
    "Order_ID": f"ORD{i:05d}",
    "Order_Date": order_date.strftime("%Y-%m-%d"),
    "Ship_Date": ship_date.strftime("%Y-%m-%d"),

    "Year": order_date.year,
    "Month": order_date.strftime("%B"),
    "Quarter": f"Q{((order_date.month-1)//3)+1}",

    "Customer_ID": f"CUST{random.randint(1001,9999)}",
    "Customer_Name": random.choice(customer_names),
    "Segment": random.choice(segments),
    "Country": "India",
    "State": state,
    "City": city,
    "Region": regions[state],
    "Category": category,
    "Sub_Category": sub_category,
    "Product_Name": sub_category,

    "Sales": sales,
    "Quantity": quantity,
    "Discount": discount,
    "Profit": profit,

    "Delivery_Days": delivery_days,
    "Profit_Margin": profit_margin,

    "Payment_Mode": random.choice(payment_modes),
    "Ship_Mode": random.choice(ship_modes)
})
# Create DataFrame
df = pd.DataFrame(data)

# Save CSV
df.to_csv("sales.csv", index=False)

print("sales.csv created successfully!")
print(df.head())
print("\nTotal Rows:", len(df))
```
